# Decision Trees

1. Start at the root node.
2. Ask a yes-or-no questions.
3. Branch based on the answer.
4. Reach the final outcome.

```python
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
X,y = make_moons(n_samples=1000, noise=0.25, random_state=42)
plt.figure(figsize=(5,4))
plt.scatter(X[:,0],X[:,1],c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title("make_moons dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

### Train a Decision Tree model

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=1)
tree.fit(X,y)
```

### Visualize the tree 

```python
plt.figure(figsize=(14,6))
plot_tree(tree, filled=True, feature_names=['Featrue 1', 'Feature 2'], class_names=['Class 0', 'Class 1'])
plt.title("Decision Tree Visualization")
plt.show()
```
