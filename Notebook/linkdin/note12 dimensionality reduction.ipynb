# Dimensionality Reduction
A technique to simplify data by keeping essential information and removing excess

# PSA
this finds key patterns to reduce data complexity effectively.

```python
print('Hello')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)
df = iris.frame.copy()
df.head()
```

```python
feature_cols = iris.feature_names
X = df[feature_cols]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:5]
```

```python
X_scaled.shape
```

### Fit PCA to reduce dimensionality

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # reduce 4 deiemsions to 2
X_pca = pca.fit_transform(X_scaled)
X_pca[:5]
```

```python
X_pca.shape
```

### Explained variance of components

```python
pca.explained_variance_ratio_
```

### Visualize the transformed data

```python
plt.figure(figsize=(6,5))
plt.scatter(X_pca[:,0],X_pca[:,1], c=df["target"], cmap=plt.cm.Set1, edgecolors='k')
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("PCA on Iris (2 components)")
plt.show()
```

> reducing dimensions speeds up analysis and reveals your data's true sttructure
