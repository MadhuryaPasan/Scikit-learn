```python
import matplotlib.pyplot as plt 
from sklearn.datasets import make_moons
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
X,y = make_moons(n_samples=1000, noise=0.25, random_state=42)

plt.figure(figsize=(5,4))
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title("Make_moons dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)

# n_estimators=1: represent single decision tree
forest = RandomForestClassifier(n_estimators=50, random_state=1)
forest.fit(X_train,y_train)

y_pred = forest.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy:.3f}")
```

### Examine feature importance

```python
import pandas as pd

feature_importance = pd.Series(forest.feature_importances_,index=['Feature 1', 'Feature 2'])
feature_importance.plot(kind='bar', color='teal', title='feature_importance')
plt.ylabel("Importance")
plt.show()

feature_importance
```
