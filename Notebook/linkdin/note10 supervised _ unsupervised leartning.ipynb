# Supervised Learning
A method where models learn from examples with known outcomes to make predictions

### Predicting with precision
- use regression for numbers
- use classification for categories


# Unsupervised Learning
A method where model find pattarans in data without predifined labels or targets.

 - Clustering groups customers by behavior
 - Dimensionality reduction reveals hidden patterns
 - Anomaly detection highlight the point that don't fit

#### Stack of papers
- supervised model predicts the label on each paper
- unsupervised model sort the papers by pattern
